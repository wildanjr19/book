12.3 Stemming using NLTK

Stemming adalah menghilangkan kata imbuhan dalam kata. seperi er, ing, be, meng. sederhananya merubah kata menjadi kata dasarnya saja.

Porter Stemmer, metode sederhana dengan menghilangkan commong ending dalam kata untuk menjadi aturan umum


```python
# import porter stemmer
from nltk.stem.porter import PorterStemmer as p_stem
```

```python
# try to sterm words
p_stem().stem('computer')
```

```python
# try abt 'dogs'
p_stem().stem('dogs')
```

```python
# lets work in words in plural form (jamak)
w_plu = ['dogs', 'egos', 'generous', 'jungles', 'natives']
```

```python
w_sgl = [p_stem().stem(wplu) for wplu in w_plu]
print(' '.join(w_sgl))
```

Snowball Stemmer

```python
# import snowball stemmer
from nltk.stem.snowball import SnowballStemmer as s_stem
```

```python
# melihat bahasa apa saja yang didukung snowball untuk stemming
print(s_stem.languages)
```

```python
# assign stemmer to use english
s_stem_eng = s_stem(language='english')
```

```python
# apply snowball stemmer (englis) to w_plu
sgls = [s_stem_eng.stem(wplu) for wplu in w_plu]
print(' '.join(sgls))
```

12.4 Stop-Words Removal with NLTK

Stop-Words adalah proses menghilangkan kata-kata yang tidak penting atau tidak dibutuhkan

```python
# import NLTK stop-words as wstops
from nltk.corpus import stopwords as wstops
```

```python
print(wstops.words('indonesian'))
```

```python
# melihat bahasa yang didukung oleh stopwords
print(wstops.fileids())
```

```python
# try with some texts
# use english stopwords
wstops_eng = wstops.words('english')
```

```python
# make some utt
utterance = 'Try to test for the stop word remove function to see how it works'
```

```python
# lets stop words the utt
utterance_clean = [word for word in utterance.split() if word not in wstops_eng]
```

fungsi diatas menggunakan 'list comprehension', jadi iterasi berdasar utterance.split(). utterance.split() akan meng-spllit kalimat menjadi kata-kata. lalu jika kata-kata yang ada dalam utterance tidak ada dalam wstops_eng maka masukkan itu ke variabel utterance_clean

```python
# display the result
utterance_clean
```

```python
# try with gutenberg library
from nltk.corpus import  gutenberg as gub
```

```python
# use txt from gutenberg
hamlet = gub.words('shakespeare-hamlet.txt')
```

```python
# lets clean the hamlet with wstops_eng
# create new var
hamlet_clean = [word for word in hamlet if word not in wstops_eng]
```

12.5 Text analysis with NLTK

Ketika data sudah diproses dan di tokenisasi, analisis dasar yang dibutuhkan adalah menghitung kata atau token, disribusinya dan frekuensi.

```python
# import webtext as wbtxt
from nltk.corpus import webtext as wbtxt
```

```python
# create a sample webtext
wbtxt_s = wbtxt.sents('firefox.txt')
wbtxt_w = wbtxt.words('firefox.txt')
```

```python
# review total words in firefox.txt
len(wbtxt_w)
```

```python
# see the vocab
# define vocabulary object
vocab = set(wbtxt_w)
```

```python
# displat the size of vocab
len(vocab)
```

```python
# lets see the frequency 
# use nltk.FreqDist
fdist = nltk.FreqDist(wbtxt_w)
```

```python
# see the freq word but exlude the stopwords such as the and not
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
lwords = dict([(k,v) for k,v in fdist.items() if len(k) > 3])
filtered_fdist = nltk.FreqDist(lwords)
filtered_fdist.plot(50, cumulative=False)
```

```python
# lets try with holmes
holmes_doc = open('Adventures_Holmes.txt', 'r', encoding='utf-8').read()
```

```python
# tokenize the holmes
from nltk.tokenize import word_tokenize as w_tok
holmes_doc_tok = w_tok(holmes_doc)
```

```python
# lets clean the holmes with wstops_eng
holmes_doc_clean = [word for word in holmes_doc_tok if word not in wstops_eng]
```

```python
# frequncy distribution
holmes_doc_clean_dist = nltk.FreqDist(holmes_doc_clean)
holmes_doc_clean_dist.plot(50, cumulative=False)
```

12.7 POS Tagging with NLTK

Pengkategorian kata dalam sebuah kalimat menjadi syntatik atau grammatical. sederhananya POS Tagging adalah mengidentifikasi setiap kata atau token. untuk panduan lengkap mengenai kategori dari POS dapat dilihat melalui Penn Treebank Tagset

```python
# import word tokenize and pos tag
from nltk.tokenize import  word_tokenize as w_tok
from nltk import pos_tag as p_tag
```

```python
# create and tokenizer two sample utt
utt1 = w_tok('Give me a call')
utt2 = w_tok('Call me later')
```

```python
# review pos tag of utt1
p_tag(utt1, tagset='universal')
```

```python
# review pos tag of utt2
p_tag(utt2, tagset='universal')
```

```python
# create new sample utt
utt_untag = w_tok('My dad was born in South America')
utt_untag
```

```python
utt_tagged = p_tag(utt_untag)
utt_tagged
```

```python
# let see with visualization with tree
import svgling

# import nltk.ne_chunk as chunk
from nltk import ne_chunk as chunk

# diplay POS tag chunk
chunk(utt_tagged)
```
